```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import matplotlib.pyplot as plt

data = pd.read_csv("customer_data.csv")

numerical_features = ['Age', 'Time on App (minutes)', 'Time on Web (minutes)', 'Annual Income (K$)', 'Spending Score']
data[numerical_features] = data[numerical_features].fillna(data[numerical_features].mean())


categorical_features = ['Gender', 'Membership', 'Favorite Product Base', 'Demography']
data[categorical_features] = data[categorical_features].fillna('U')
features = numerical_features + categorical_features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[numerical_features])
pca = PCA(n_components=2)
pca_result = pca.fit_transform(data_scaled)
plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1])
plt.title('PCA Result')
plt.xlabel('jbfjb 1')
plt.ylabel('bkkjrbkj  1')
plt.show()
kmeans = KMeans(n_clusters=3)
kmeans.fit(data_scaled)
clusters = kmeans.predict(data_scaled)


plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=clusters, cmap='viridis')
plt.title('Kknrn')
plt.xlabel('m fvkjn')
plt.ylabel('bjkkj')
plt.colorbar(label='Cluster')
plt.show()

data_encoded = pd.get_dummies(data[categorical_features])


final_data = pd.concat([pd.DataFrame(pca_result, columns=['PC1', 'PC2']),
                        pd.DataFrame(clusters, columns=['Cluster']),
                        data_encoded], axis=1)


print(final_data)
```
